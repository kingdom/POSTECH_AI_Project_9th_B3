```python
def test_image(folder):
    imgs = []
    cctv_names = []
    for name in os.listdir(dir+folder):
        if name == '.ipynb_checkpoints':
            continue
        cctv_names.append(name[:-4])
        imgs.append(img_crop(dir+folder+name,'img'))

    addv_1 = cv2.hconcat([imgs[0],imgs[1]])
    addv_2 = cv2.hconcat([imgs[2],imgs[3]])
    res = cv2.vconcat([addv_1,addv_2])

    now = time.strftime('%Y%m%d%H%M%S', time.localtime(time.time()))
    a,b,c,d = cctv_names
    print(cctv_names)
    cv2.imwrite(f'4windows_input/{now}_{a}_{b}_{c}_{d}.jpg', res)
```

```python
def make_3_data(n):
    for i in range(n): # 생성 윈도우의 개수
        test_image('melt_factory3/')
        test_image('melt_factory4/')
        test_image('melt_factory1/')
        time.sleep(1)
        print()
        test_image('melt_factory3/')
        test_image('melt_factory4/')
        test_image('melt_factory2/')
        time.sleep(1)
        print()
    print('예제 데이터 종료')
```
